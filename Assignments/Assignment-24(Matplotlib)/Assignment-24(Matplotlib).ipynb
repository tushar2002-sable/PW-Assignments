##### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a data visualization library in Python used for creating static, animated, and interactive visualizations. It is widely used for data analysis and presentation. The Pyplot module in Matplotlib provides a simple interface for creating various types of plots. Here are five types of plots that can be created using Pyplot:

1. Line Plot: Displays data points connected by straight lines.
2. Bar Plot: Represents data using rectangular bars of different heights.
3. Scatter Plot: Shows the relationship between two variables using individual data points.
4. Histogram: Visualizes the distribution of a dataset by grouping data into bins.
5. Pie Chart: Displays data as a circular chart divided into slices, showing the proportion of each category.

These plots, along with many others, can be customized with labels, titles, colors, and more to effectively visualize and analyze data using Matplotlib's Pyplot module.

##### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
```
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```

A scatter plot is a type of data visualization that represents the relationship between two variables. It displays data points as individual dots on a two-dimensional graph, where each dot represents the values of the variables.

````python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.scatter(x,y)
````

##### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
```
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

The `subplot()` function in Matplotlib is used to create a grid of subplots within a single figure. It allows us to divide the figure into multiple smaller plots or axes, which can be useful for displaying multiple visualizations side by side or for creating complex layouts.

The `subplot()` function takes three arguments: the number of rows, the number of columns, and the index of the current subplot. These arguments determine the layout of the grid and specify the position of the current subplot within the grid.

````python
# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("Line 1")

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("Line 2")

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("Line 3")

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("Line 4")

# Adjust spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
````

##### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
```
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A bar plot is a type of data visualization that uses rectangular bars to represent and compare categorical data. It is commonly used to display the distribution, comparison, or relationship between different categories or groups.

Bar plots are used for the following purposes:

1. Comparison: Bar plots allow for a visual comparison of values between different categories or groups. The length or height of each bar represents the quantity or frequency associated with that category, making it easy to identify differences or similarities.

2. Distribution: Bar plots can illustrate the distribution of data across categories. They provide a clear representation of how data points are spread among various groups, allowing for a quick understanding of the distribution patterns.

3. Relationship: Bar plots can showcase the relationship between two or more variables. By comparing the heights of bars for different categories across variables, it is possible to observe correlations, trends, or patterns.

4. Communication: Bar plots are widely used for data presentation and communication. They offer a concise and visually appealing way to convey information to an audience, making it easier to understand and interpret categorical data.

````python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.xlabel("Profit(in Millions)")
plt.ylabel("Companies")
plt.title("Profit of Companies")
plt.barh(company, profit)
plt.show()
````

##### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
```
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of a data distribution. It is used to display the distribution of a dataset, identify outliers, and compare multiple datasets or groups.

Box plots are particularly useful for the following purposes:

1. Distribution Summary: A box plot provides a visual summary of the distribution of a dataset, including measures such as the median, quartiles, and range.

2. Outlier Detection: Box plots can identify potential outliers, which are data points that significantly deviate from the rest of the distribution. Outliers are represented as individual data points beyond the whiskers of the plot, helping to highlight extreme values.

3. Comparison of Datasets: Box plots allow for easy comparison of multiple datasets or groups. By placing multiple box plots side by side, it becomes simple to compare their medians, quartiles, and ranges, providing insights into differences or similarities between the groups.

4. Skewness and Symmetry: Box plots provide visual cues about the skewness and symmetry of a dataset. The shape of the box and whiskers can indicate whether the distribution is symmetrical or skewed to one side.

````python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

dataset = [box1, box2]
plt.xlabel("Box")
plt.ylabel("Values")
plt.title("Box Plot Example")
plt.boxplot(dataset)

plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
````
