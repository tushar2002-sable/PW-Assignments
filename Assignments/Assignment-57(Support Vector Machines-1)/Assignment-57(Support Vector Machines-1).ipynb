### Q1. What is the mathematical formula for a linear SVM?

The mathematical formula for a linear support vector machine (SVM) is:

```
w⋅x+b=0
```

where:

* **w** is the weight vector
* **x** is the input vector
* **b** is the bias term

The goal of the SVM is to find the weight vector **w** and bias term **b** that maximize the margin between the two classes. The margin is the distance between the hyperplane and the nearest data points of each class.

### Q2. What is the objective function of a linear SVM?

The objective function of a linear support vector machine (SVM) is a quadratic programming (QP) problem that can be formulated as follows:

```
min 12w2+C∑i=1(max(0,1−yi(w⋅xi+b)))
```

where:

* **w** is the weight vector
* **C** is a hyperparameter that controls the trade-off between the margin and the misclassification errors
* **yi** is the label of the i-th data point
* **xi** is the i-th data point

The goal of the SVM is to find the weight vector **w** and bias term **b** that minimizes the objective function while also satisfying the following constraints:

* **w⋅xi+b≥1** for all i such that yi=1
* **w⋅xi+b≤−1** for all i such that yi=−1

The first set of constraints ensures that the hyperplane separates the positive and negative data points with a margin of at least 1. The second set of constraints ensures that all of the data points are classified correctly.

The hyperparameter **C** controls the trade-off between the margin and the misclassification errors. A larger value of C will result in a larger margin, but it will also increase the risk of misclassification errors. A smaller value of C will result in a smaller margin, but it will also decrease the risk of misclassification errors.

The SVM can be solved using a variety of methods, such as the interior point method or the gradient descent method.

### Q3. What is the kernel trick in SVM?

The kernel trick is a technique used in support vector machines (SVMs) to map data points from a lower-dimensional space to a higher-dimensional space. This mapping is done using a kernel function. The kernel function is a mathematical function that measures the similarity between two data points.

The kernel trick allows SVMs to be used for classification and regression tasks even when the data points are not linearly separable in the original space. This is because the kernel function can map the data points to a higher-dimensional space where they become linearly separable.

There are many different kernel functions that can be used with SVMs. Some of the most common kernel functions include:

* The linear kernel: This is the simplest kernel function. It measures the similarity between two data points using the dot product.
* The polynomial kernel: This kernel function measures the similarity between two data points using a polynomial of a certain degree.
* The radial basis function (RBF) kernel: This kernel function measures the similarity between two data points using an exponential function.

````python
# Load the Iris dataset
iris = load_iris()

# Split the dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
````

````python
# Compute the accuracy of the model on the testing set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
````

````python
def linear_svm(X, y, C):
    # Initialize the weight vector w and the bias term b.
    w = np.zeros(X.shape[1])
    b = 0

    # Optimize the weight vector w and the bias term b using gradient descent.
    for i in range(len(X)):
        loss = y[i] * np.dot(w, X[i]) + b
        grad_w = y[i] * X[i]
        grad_b = y[i]
        
        w -= C * grad_w
        b -= C * grad_b

    return w, b

def predict(X, w, b):
    scores = np.dot(X, w) + b
    predictions = np.sign(scores)

    return predictions
````
