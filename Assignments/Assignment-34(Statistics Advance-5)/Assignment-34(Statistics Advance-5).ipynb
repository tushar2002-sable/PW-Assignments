##### Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
import scipy.stats as stats

expected_distribution = [20, 20, 20, 10, 10, 20]
observed_distribution = [15, 20, 17, 14, 16, 18]

chi_squared, p_value = stats.chisquare(observed_distribution, expected_distribution)
print("Chi_squared Value: ", chi_squared)

## find the critical value
significance_vlaue = 0.05
df = len(expected_distribution) - 1
critical_value = stats.chi2.ppf(1 - significance_vlaue, df)
print("Critical Value: ", critical_value)

if chi_squared > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")

"""We can conclude that the observed distribution is a good fit for the expected distribution."""
```

##### Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following data: Interpret the results of the test.


```python
import numpy as np
import scipy.stats as stats

observed_data = np.array([[20, 15],  # Row 1
                          [10, 25],  # Row 2
                          [15, 20]]) # Row 3

# Calculate the chi-squared statistic and p-value
chi_squared, p_value, dof, expected = stats.chi2_contingency(observed_data)

# Set the significance level (alpha)
significance_level = 0.05

# Find the critical value
critical_value = stats.chi2.ppf(1 - significance_level, dof)

# Print the chi-squared statistic, p-value, and critical value
print("Chi-squared statistic:", chi_squared)
print("P-value:", p_value)
print("Critical value:", critical_value)

# Compare chi-squared statistic with the critical value
if chi_squared > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

##### Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

```python
import statsmodels.stats.proportion as proportion

# Given data
sample_size = 500
sample_smokers = 60

# Calculate the 95% confidence interval using statsmodels.stats.proportion.proportion_confint
confidence_interval = proportion.proportion_confint(sample_smokers, sample_size, alpha=0.05, method='normal')

# Print the results
print("95% Confidence Interval for the proportion of smokers:")
print(f"Lower Bound: {confidence_interval[0]:.4f}")
print(f"Upper Bound: {confidence_interval[1]:.4f}")
```

##### Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Degrees of freedom
df = 10

# Range of chi-square values for the plot
x = np.linspace(0, 30, 500)

# Calculate the chi-square distribution
chi_squared = stats.chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, chi_squared, label='Chi-square distribution with 10 degrees of freedom')

# Shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(15, 30, 500)
y_shade = stats.chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, color='grey', alpha=0.5, label='Area for chi-square stat of 15')

# Label the axes and add a legend
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability Density Function')
plt.legend()

# Show the plot
plt.show()
```

##### Q9. A study was conducted to determine if there is an association between smoking status (smoker or non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between smoking status and lung cancer diagnosis. Use a significance level of 0.05.


```python
import scipy.stats as stats
import numpy as np

observed = np.array([[60,140],  # Row 1
                     [30,170]]) # Row 2

chi_square, p_value, dof, expected = stats.chi2_contingency(observed)

# Print the results
print("Chi-square statistic:", chi_square)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)

# Interpret the results based on the p-value and significance level (alpha)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Accept the null hypothesis. There is no significant association between smoking status and lung cancer diagnosis.")
```

##### Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

```python
import scipy.stats as stats
import numpy as np

sample_size = 30
sample_mean = 72
sample_std = 10
population_mean = 70
alpha = 0.05

t_statistics = sample_mean - population_mean/ (sample_std /(np.sqrt(sample_size)))
print(f"t_statistics: {t_statistics}")

df = sample_size - 1

critical_t_value = stats.t.ppf(1 - alpha/2, df)
print(f"critical value: {critical_t_value}")

if t_statistics > critical_t_value:
    print("Reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("Accept the null hypothesis. There is no significant difference between the population mean and 70.")
```
