##### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Sure, here are five plots that we can plot using the Seaborn library, along with their uses:

**1. ** **Barplot:** A barplot is a type of chart that is used to show the distribution of data in categories. It is often used to compare different categories of data.

**Use:** Barplots are used to show the frequency of categorical data. They can also be used to show the relative size of different categories.

**2. ** **Lineplot:** A lineplot is a type of chart that is used to show the relationship between two variables over time. It is often used to show how a variable changes over time.

**Use:** Lineplots are used to show the trend of a variable over time. They can also be used to show the correlation between two variables.

**3. ** **Scatterplot:** A scatterplot is a type of chart that is used to show the relationship between two variables. It is often used to show how two variables are correlated.

**Use:** Scatterplots are used to show the relationship between two variables. They can also be used to identify outliers in the data.

**4. ** **Histogram:** A histogram is a type of chart that is used to show the distribution of data in a continuous range. It is often used to show the distribution of a single variable.

**Use:** Histograms are used to show the distribution of a single variable. They can also be used to identify the shape of the distribution.

**5. ** **Heatmap:** A heatmap is a type of chart that is used to show the relationship between two variables in a matrix. It is often used to show the correlation between two variables.

**Use:** Heatmaps are used to show the relationship between two variables in a matrix. They can also be used to identify clusters of data.

##### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
```Note: timepoint, signal, event, and region are columns in the fmri dataset.```

````python
import seaborn as sns

fmri = sns.load_dataset('fmri')
sns.lineplot(x=fmri.timepoint, y=fmri.signal, data=fmri, hue="event", style="region")
````

##### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
```Note: pclass, age, and fare are columns in the titanic dataset.```

````python
titanic = sns.load_dataset('titanic')
sns.boxplot(x='pclass', y='age', data=titanic)
````

````python
sns.boxplot(x='pclass', y='fare', data=titanic)
````

##### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

````python
diamonds = sns.load_dataset('diamonds')
sns.histplot(data=diamonds, x='price', hue='cut')
````

##### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

````python
iris = sns.load_dataset('iris')
sns.pairplot(data=iris, hue='species')
````

##### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

````python
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(data=flights_pivot, annot=True, cmap="Blues", fmt='d')
````
